### AGE VARIABLE

```python
# Creating age category column
bank_add_full.age.describe()
```

```python
#min is 17 and max is 98
#Creating age category
groups = bank_add_full['age'].apply(lambda x: '17-25' if x<=25 else ('26-40' if x<=40 else ('41-60' if x<=60 else '60+')))
bank_add_full['age_group'] = groups

age_group_counts = pd.DataFrame(bank_add_full.groupby('age_group').age.count()).rename({'age':'count'}, axis = 1).reset_index()
print(age_group_counts)
plt.figure(figsize=(10,6))
sns.barplot(x = 'age_group', y = 'count', data = age_group_counts)
plt.title('Classification of the customer base across age-groups',fontdict={'color':'red','size':15})
plt.ylabel('Number of customers',fontdict={'color':'green','size':12})
plt.xlabel('age-group',fontdict={'color':'green','size':12})
plt.show()
```

```python
# 'y' and 'age-group'
# Looking at relation between 'y' and different age-groups
age_y = pd.DataFrame(bank_add_full.groupby('age_group').y.value_counts()).rename({'y':'count'}, axis = 1).reset_index()

plt.figure(figsize = (10,7))
plt.title('Age-wise classification of customers against y',fontdict={'color':'red','size':20})
sns.barplot(x = 'age_group', y = 'count', hue = 'y', data = age_y)
plt.ylabel('Count',fontdict={'color':'green','size':12})
plt.xlabel('age-group',fontdict={'color':'green','size':12})
plt.show()
```

- In the age-groups of 26-40 and 41-60 yrs, majority of the people are not subscribed to the term deposit plan
